## Transform Dataframes

```python
# Create filtered data 
rpm_cols = ["NAME", "GP", "ORPM", "DRPM", "RPM", "WINS"]
rpm_transformed = rpm_df[rpm_cols].copy()
name_split = rpm_transformed["NAME"].str.split(",", n = 1, expand = True) 
rpm_transformed["NAME"] = name_split[0]

#Rename Columns
rpm_transformed = rpm_transformed.rename(columns={"NAME": "player_name",
                                                  "GP": "player_games_played",
                                                  "ORPM": "player_off_plus_min",
                                                  "DRPM": "player_def_plus_min",
                                                  "RPM": "player_real_plus_min",
                                                  "WINS": "player_wins" })


rpm_transformed["player_name"] = rpm_transformed["player_name"].str.lower()
rpm_transformed["player_name"] = rpm_transformed["player_name"].str.replace(" ", "")
rpm_transformed.set_index("player_name", inplace=True)

rpm_transformed.head(10)
```

```python
pie_cols = ["PLAYER", "TEAM", "NETRTG", "EFG%", "TS%","USG%", "PIE"]
pie_transformed = pie_df[pie_cols].copy()

pie_transformed = pie_transformed.rename(columns={"PLAYER": "player_name",
                                                  "TEAM": "player_team",
                                                  "NETRTG": "player_net_rating",
                                                  "EFG%": "player_eff_fg_pct",
                                                  "TS%": "player_true_shooting_pct",
                                                  "USG%": "player_usage_rate",
                                                  "PIE":"player_impact_estimate"})


# pie_transformed['player_name'] = pie_transformed['player_name'].astype(str)
pie_transformed["player_name"] = pie_transformed["player_name"].str.lower()

pie_transformed["player_name"] = pie_transformed["player_name"].str.replace(" ", "")
pie_transformed.set_index("player_name", inplace=True)
pie_transformed.head(20)
```

```python
# TYLER's Data for Player Attributes
attributes = pd.read_csv('Resources/all_seasons.csv')
attributes = attributes[['player_name', 'team_abbreviation', 'age', 'player_height', 'player_weight', 'college', 'draft_year', 'season']]
attributes = attributes.loc[attributes['season'] == '2016-17']
attributes.reset_index(inplace=True)
attributes['season'] = attributes['season'].str.split('-')[0][0]
attributes['player_name'] = attributes['player_name'].str.lower()
attributes['player_name'] = attributes['player_name'].str.replace(" ", "")
attributes.drop(columns=['index'], inplace=True)
attributes = attributes.set_index('player_name')
attributes.head()
```

```python
#Marcus Data Transform 

salary_file = "Resources/nba_2017_salary.csv"
salary_df = pd.read_csv(salary_file)


twitter_file = "Resources/nba_2017_twitter_players.csv"
twitter_df = pd.read_csv(twitter_file)



salary_df = salary_df.rename(columns={"NAME": "player_name",
                                     "POSITION": "player_position",
                                     "TEAM": "player_team",
                                     "SALARY": "player_salary"})

twitter_df = twitter_df.rename(columns={"PLAYER": "player_name",
                                     "TWITTER_FAVORITE_COUNT": "favorite_count",
                                     "TWITTER_RETWEET_COUNT": "retweet_count"})

salary_df['player_name'] = salary_df['player_name'].str.replace(" ","")
salary_df['player_name'] = salary_df['player_name'].str.replace(".","")
salary_df['player_name'] = salary_df['player_name'].str.replace("-","")
salary_df['player_name'] = salary_df['player_name'].str.lower()



twitter_df['player_name'] = twitter_df['player_name'].str.replace(" ","")
twitter_df['player_name'] = twitter_df['player_name'].str.replace(".","")
twitter_df['player_name'] = twitter_df['player_name'].str.replace("-","")
twitter_df['player_name'] = twitter_df['player_name'].str.lower()

twitter_df.set_index("player_name", inplace=True)
salary_df.set_index("player_name", inplace=True)   

salary_df.head(10)
```

```python
#Extract CSVs into DataFrames
nba_2017_players_stats_combined_file = "Resources/nba_2017_players_stats_combined.csv"
nba_player_df = pd.read_csv(nba_2017_players_stats_combined_file)
nba_player_df.dropna(inplace=True, how='all')

nba_player_df.to_csv("CSVs/player_stats_transformed.csv")
```

```python
#Transform premise DataFrame
# Create a filtered dataframe from specific columns
nba_2017_stats_cols = ["PLAYER", "POSITION", "TEAM", "POINTS", "TRB", "AST", "STL", "BLK", "FG%", "3P%", "FT%"]
nba_2017_stats_transformed = nba_player_df[nba_2017_stats_cols].copy()

# Rename the column headers
nba_2017_stats_transformed = nba_2017_stats_transformed.rename(columns={"PLAYER": "player_name",\
                                                                        "POSITION": "player_position",
                                                                        "TEAM": "player_team_abbreviation",
                                                                        "POINTS": "player_pts",
                                                                        "TRB": "player_trb",
                                                                        "AST": "player_ast",
                                                                        "STL": "player_stl",
                                                                        "BLK": "player_blk",
                                                                        "FG%": "player_fg_pct",
                                                                        "3P%":"player_threept_pct",
                                                                        "FT%":"player_ft_pct"})


nba_2017_stats_transformed["player_name"] = nba_2017_stats_transformed["player_name"].str.lower()
nba_2017_stats_transformed["player_name"] = nba_2017_stats_transformed["player_name"].str.replace(" ", "")
# Clean the data by dropping duplicates and setting the index
nba_2017_stats_transformed.drop_duplicates("player_name", inplace=True)
nba_2017_stats_transformed.set_index("player_name", inplace=True)

nba_2017_stats_transformed.head()
```
